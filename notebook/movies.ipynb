```python
import numpy as np 
import pandas as pd
```

```python
movies = pd.read_csv('tmdb_5000_movies.csv', encoding='latin-1')
credits = pd.read_csv('tmdb_5000_credits.csv', encoding='latin-1') 
```

```python
movies.head()
```

```python
movies.shape
```

```python
credits.head()
```

```python
movies = movies.merge(credits,on='title')
```

```python
movies.head()
```

```python
movies = movies[['movie_id','title','overview','genres','keywords','cast','crew']]
```

```python
movies.head()
```

```python
import ast
```

```python
#function to convert the dictionaries into the list
def convert(text):
    L = []
    for i in ast.literal_eval(text):
        L.append(i['name']) 
    return L 
```

```python
movies.dropna(inplace=True)
```

```python
movies['genres'] = movies['genres'].apply(convert)
movies.head()
```

```python
movies['keywords'] = movies['keywords'].apply(convert)
movies.head()
```

```python
#we want only top 3 cast names...function to get only top 3 cast names
def convert3(text):
    L = []
    counter = 0
    for i in ast.literal_eval(text):
        if counter < 3:
            L.append(i['name'])
        counter+=1
    return L 
```

```python
movies['cast'] = movies['cast'].apply(convert)
movies.head()
```

```python
movies['cast'] = movies['cast'].apply(lambda x:x[0:3])
```

```python
#among all the crew only director name is required...function to get only name of the director
def fetch_director(text):
    L = []
    for i in ast.literal_eval(text):
        if i['job'] == 'Director':
            L.append(i['name'])
            break 
    return L 
```

```python
movies['crew'].apply(fetch_director)
```

```python
movies['crew'] = movies['crew'].apply(fetch_director)
```

```python
movies.head()
```

```python
#removing blank space between words
def collapse(L):
    L1 = []
    for i in L:
        L1.append(i.replace(" ",""))
    return L1
```

```python
movies['cast'] = movies['cast'].apply(collapse)
movies['crew'] = movies['crew'].apply(collapse)
movies['genres'] = movies['genres'].apply(collapse)
movies['keywords'] = movies['keywords'].apply(collapse)
```

```python
#converting overview column into a list
movies['overview'] = movies['overview'].apply(lambda x:x.split())
```

```python
#combining overview, genres, keywords, cast, crew into one tags column
movies['tags'] = movies['overview'] + movies['genres'] + movies['keywords'] + movies['cast'] + movies['crew']
```

```python

new = movies.drop(columns=['overview','genres','keywords','cast','crew'])
```

```python
new.head()
```

```python
new['tags'] = new['tags'].apply(lambda x: " ".join(x))
new.head()
```

```python
import nltk
```

```python
def stem(text):
    y = []

    for i in text.split():
        y.append(ps.stem(i))
    
    return " ".join(y)
```

```python
from nltk.stem.porter import PorterStemmer
ps = PorterStemmer()
```

```python
new['tags'] = new['tags'].apply(stem)
```

```python
#using bag of words, top 5000 words extracting
from sklearn.feature_extraction.text import CountVectorizer
cv = CountVectorizer(max_features=5000,stop_words='english')
```

```python
vector = cv.fit_transform(new['tags']).toarray()
```

```python
vector.shape
```

```python
new.head()
```

```python
#now to get similarity between movies, using cosine similarity to find the minimum distance between vector for similar movies
from sklearn.metrics.pairwise import cosine_similarity
```

```python
similarity = cosine_similarity(vector)
```

```python
similarity[0]
```

```python
def recommend(movie):
    index = new[new['title'] == movie].index[0]
    distances = sorted(list(enumerate(similarity[index])),reverse=True,key = lambda x: x[1])
    for i in distances[1:6]:
        print(new.iloc[i[0]].title)
```

```python
recommend('Iron Man')
```
